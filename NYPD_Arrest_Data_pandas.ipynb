# Selecting/ Filtering data

```python
#select the condition we want to see

select_condition = df["age_group"] == "18-24"

select_condition
```

```python
# making a data frame to only display age group of 18-24
df_age =  df[select_condition]
df_age.head()
```

```python
# quick check for dupes and nulls 
df.isnull()
df.isnull().sum().sum() # if there is any nulls, the number will not be 0
```

```python
# drop the nulls 
df = df.dropna() # no more nulls
```

```python
df.duplicated().sum() # no dupes, 

# to drop duplicates
# df = df.drop_duplicates()
```

```python
# check if there is any dates before 2021
date_condition = df["arrest_date"] < "01/02/2021"
df_date = df[date_condition]
df_date.head()
```

```python
# two conditions at once 
con_1 = df["arrest_boro"] == "Q"
con_2 = df["age_group"] == "<18"

df_mult_con = df[con_1 & con_2]
df_mult_con.head()
```

```python
# select conditions using "or"(|)
con_1 = df["arrest_boro"] == "K"
con_2 = df["arrest_boro"] == "B"

df_new =df[con_1 | con_2]

df_new.head()
```

```python
# using .isin to select through the police precinct
list_of_precinct = [75, 42, 101]
select_statement = df.arrest_precinct.isin(list_of_precinct)

df[select_statement]
```

```python
# use ~ to look for not the precinct 
df[~select_statement]
```

```python
# using .where
# .where will take a condition, and values for when the statement is true or false 
c1 = df["perp_race"] == "BLACK HISPANIC"

df["is_black_hispanic"] = np.where(c1,"is black hispanic", "not black hispanic")


select_statement = df["is_black_hispanic"] == "is black hispanic"

df_new_statement = df[select_statement]

df_new_statement.head()

```

# Using groupby

```python
gb = df.groupby("perp_race")
gb["perp_race"].agg(["count"])
```

```python
def race_shape (race):
    x = race 
    select_race = df["perp_race"] == x
    df_new_race = df [select_race]
    print(race, " shape ", df_new_race.shape )
```

```python
race_list = ["AMERICAN INDIAN/ALASKAN NATIVE", "ASIAN / PACIFIC ISLANDER", "BLACK", "BLACK HISPANIC", "UNKNOWN", "WHITE", "WHITE HISPANIC"]

for race in race_list:
    race_shape(race)
```

```python
gb_sex = df.groupby("perp_sex")
gb_sex["perp_sex"].agg(["count"])
```

```python
# making a dataframe from a groupby 

gb_boro_race = df.groupby("arrest_boro")["perp_race"].count()
new_df = pd.DataFrame(gb_boro_race)

new_df = new_df.reset_index()

new_df
```
